```python
# Import packages
import pandas as pd
import datetime as dt
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import timedelta
from datetime import datetime
```

```python
# Import data
price_history = pd.read_csv(
    'data_price_history.csv', sep='|', index_col=['yahoo_code']).drop_duplicates()
price_history['Date'] = pd.to_datetime(price_history['Date'])
price_history.head()
```

```python
stock_info = pd.read_csv('data_stock_info.csv', sep='|',
                         index_col='yahoo_code').drop_duplicates()
stock_info.head()
```

```python
df = price_history.join(stock_info).reset_index().set_index('Date').sort_index()
df.head()
```

```python
df.groupby('main_city')['pct_change'].describe()
```

```python
# Clearly there are some anomolies:
df.sort_values('pct_change', ascending=False).head(10)
```

```python
analysis_days = 5
anomaly_date = '2017-02-24'
anomaly_stock = 'BAS.BR'
formatted_date = datetime.strptime(anomaly_date, '%Y-%m-%d').date()
start_date = str(formatted_date + timedelta(days=-analysis_days))
end_date = str(formatted_date + timedelta(days=analysis_days))

filter = (df.index >= start_date ) & (df.index <= end_date) & (df['yahoo_code'] == anomaly_stock)
df.loc[filter]
```
